```python
#importing some useful libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define data
x = np.array([20.34, 21.56, 21.89, 22.63, 20.78, 20.98, 20.10, 22.54, 21.23, 22.34, 19.14, 22.61, 23.01, 19.89, 20.04, 20.95, 21.09, 22.12, 21.36, 21.05, 20.56, 21.04, 22.80, 20.85, 20.34, 19.36, 18.89, 19.54, 20.01, 19.03])
```

```python
#caluclate sample mean, sample variance, and sample standard deviation
_x = x.mean()
s2 = x.var()
s = np.sqrt(s2)

print("Sample Mean: %f\nSample Variance: %f\nSample Standard Deviation %f" % (_x, s2, s))
```

```python
#find the minumum and maximum observations in the sample, and calculate the range
min_x = x.min() #minimum
max_x = x.max() #maximum
range_x = max_x - min_x #range

print("Minimum observation: %f\nMaximum observation: %f\nRange: %f" % (min_x, max_x, range_x))
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,12))

ret = ax.hist(x, bins=10, range=(min_x, max_x), density=True, edgecolor = 'black')
edges = ret[1] #bins' edges
val = np.add(edges[:-1], edges[1:])/2

ax.set_xticks(val) #show the values on the x-axis
ax.set_title('Frequency Histogram')
plt.savefig('hist.png', dpi=300)
```

```python
s = np.sqrt(s2)
fig, ax = plt.subplots(1,1, figsize=(16,12))

ret = ax.hist(x, bins=10, range=(min_x, max_x), density=True, edgecolor = 'black')
edges = ret[1] #bins' edges
val = np.add(edges[:-1], edges[1:])/2

ax.set_xticks(val) #show the values on the x-axis
ax.axvline(x = _x, color = 'red', label = 'mean')
ax.axvline(x = _x + 2*s, color ='red', linestyle = '--', label = 'two sd')
ax.axvline(x = _x - 2*s, color ='red', linestyle = '--')

ax.legend()
ax.set_title('Frequency Histogram')
plt.savefig('hist_mean.png', dpi=300)
```

```python
s = np.sqrt(s2)
fig, ax = plt.subplots(1,1,figsize=(16,12))

ret = ax.hist(x, bins=10, range=(min_x, max_x), density=True, edgecolor = 'black')
edges = ret[1] #bins' edges

heights = ret[0] #bins' height

width = edges[1] - edges[0]

edges = np.concatenate((np.ones(1)*(edges[0] - width), edges, np.ones(1)*(edges[-1] + width)))
heights = np.concatenate((np.zeros(1), heights, np.zeros(1)))

val = np.add(edges[:-1], edges[1:])/2

ax.set_xticks(val) #show the values on the x-axis
ax.axvline(x = _x, color = 'red', label = 'mean')
ax.axvline(x = _x + 2*s, color ='red', linestyle = '--', label = 'two sd')
ax.axvline(x = _x - 2*s, color ='red', linestyle = '--')

ax.plot(val, heights, label= 'rel freq polygon')

ax.legend()
ax.set_title('Frequency Histogram')
plt.savefig('hist_mean_freq_polyg.png', dpi=300)
```

```python
# Define the sample data
data = [20.34, 21.56, 21.89, 22.63, 20.78, 20.98, 20.10, 22.54, 21.23, 22.34]

# Calculate the mean (average) of the sample
mean = sum(data) / len(data)

# Calculate the squared differences from the mean
squared_diff = [(x - mean) ** 2 for x in data]

# Calculate the sample variance
sample_variance = sum(squared_diff) / (len(data) - 1)

# Print the result
print("Sample Variance:", sample_variance)
print("Sample Standard Deviation:", np.sqrt(sample_variance))
```
