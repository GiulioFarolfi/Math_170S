```python
import numpy as np
```

```python
x = np.array([22,36,65,44,41,29,52,38,34,39,27,40,30,49,31,58,75,62,47,62,103,46,53,21,18,60,25,37,42,31])
```

```python
max_x = np.max(x)
min_x = np.min(x)
avg_x = np.mean(x)
q25_x = np.quantile(x, 0.25)
q75_x = np.quantile(x, 0.75)
iqr_x = q75_x - q25_x

print("Sample Maximum: %f\nSample Minimum: %f\nSample Mean: %f\nFirst Quartile: %f\nThird Quartile: %f\nIQR: %f" % (max_x, min_x, avg_x, q25_x, q75_x, iqr_x))
```

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(1,1,figsize=(16,12))

# Create a boxplot
ax.boxplot(x, vert=False)

# Set axis labels
ax.set_xlabel('Values')
ax.set_ylabel('Data Set')

# Set the title
plt.title('Horizontal Boxplot')

# Save the plot
plt.savefig('horiz_boxplot.png', dpi=300)
```

```python
import scipy.stats as stats

# Generate data points for the standard normal distribution
x = np.linspace(-3, 3, 1000)
pdf = stats.norm.pdf(x, loc=0, scale=1)

# Define quartile values
quartiles = [0.25, 0.5, 0.75]

# Calculate quartile values in terms of z-scores
quartile_z_scores = [stats.norm.ppf(q) for q in quartiles]

# Create a figure and axis
fig, ax = plt.subplots(1,1,figsize=(16,12))

# Plot the PDF curve
ax.plot(x, pdf, label='PDF')

# Color-coded quartile lines
for z_score in quartile_z_scores:
    ax.axvline(z_score, color='red', linestyle='--', label=f'Quartile at {z_score:.2f}')

# Add labels and legend
ax.set_xlabel('Values')
ax.set_ylabel('Density')
ax.legend()

# Set the title
plt.title('Standard Normal Distribution PDF with Color-Coded Quartiles')

# Save the plot
plt.savefig('norm_pdf_quartiles.png', dpi=300)
```
